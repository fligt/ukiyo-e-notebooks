### Inkttime test

```python
import matplotlib.pyplot as plt
import numpy as np
import inktime as ink
```

```python
test_img = ink.kaleidoscope()

fig, ax = plt.subplots(figsize=[6, 6])
ax.imshow(test_img)
ax.set_xticks([])
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('kaleidoscope test image');
```

```python
ink.palette_plot(test_img)
primaries = ink.primaries(test_img)
print(primaries)
```

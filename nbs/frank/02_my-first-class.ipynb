```python
class Rekenmachine():

    def __init__(self, a, b): 
        '''geef a en b'''

        self.a = a 
        self.b = b 

    def plus_of_min(self, operation):
        '''niet nodig'''

        result = plus_of_min(self.a, self.b, operation)


        return result
```

```python
def plus_of_min(a, b, operation): 

    if operation == '+': 
        result = a + b
    if operation == '-': 
        result = a - b 
    return result
```

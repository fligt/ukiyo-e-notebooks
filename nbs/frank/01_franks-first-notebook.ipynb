*A few remarks on sharing notebooks via github.* 

The easiest way to exchange our python code in Jupyter notebooks is via this github repository. In order avoid merge conflicts it is best to make changes in our own folders. 

To start working you need open a terminal in your root folder and the clone the repo to your local machine once: 

    $ git clone git@github.com:fligt/ukiyo-e-notebooks.git (this downloads the all files and their version histories to your local computer)

When working, typically you will need to open a terminal in the root folder the following sequence of terminal commands: 

    $ git pull (this downloads the lastest files from github)  
    
    $ git add -A (this tells git to include all changes that you have made)  
    
    $ git commit -m "my first commit message" (this includes your lastest changes in the local version history of git)  
    
    $ git push (this uploads the latest changes to github)  

Good luck!
